```python
label = data[['hhb','hbo2','ca','na']]

data.drop(columns=['id','hhb','hbo2','ca','na'], inplace=True)
test.drop(columns=['id'], inplace=True)
```

```python
cat_columns = [] 
num_columns = [c for c in data.columns if c not in cat_columns]
```

```python
def preprocess(x_train, x_valid, x_test):
    
    tmp_x_train = x_train.copy()
    tmp_x_valid = x_valid.copy()
    tmp_x_test  = x_test.copy()
    
    # 결측치 처리 : 보간법 - 데이콘 코드 공유, 실무자에 따르면 스펙트럼 데이터를 단순 평균으로 대체하는 것은 큰 오류를 범할 수 있으므로 보간법으로 대체해야 한다고함
    tmp_x_train = tmp_x_train.interpolate(methods='linear', axis=1)
    tmp_x_valid = tmp_x_valid.interpolate(methods='linear', axis=1)
    tmp_x_test = tmp_x_test.interpolate(methods='linear', axis=1)     
    
    # 스케일링
    x_train_mean = np.mean(tmp_x_train[num_columns], axis=0)
    x_train_std  = np.std(tmp_x_train[num_columns], axis=0)

    tmp_x_train.loc[:, num_columns] = (tmp_x_train[num_columns] - x_train_mean) / (x_train_std + 1e-4)
    tmp_x_valid.loc[:, num_columns] = (tmp_x_valid[num_columns] - x_train_mean) / (x_train_std + 1e-4)
    tmp_x_test.loc[:, num_columns] = (tmp_x_test[num_columns] - x_train_mean) / (x_train_std + 1e-4)
    
    # 인코딩 : 전부 연속형 변수라 불필요
    
    # 리턴
    return tmp_x_train, tmp_x_valid, tmp_x_test
```

```python
tmp_x_train = x_train.copy()
tmp_x_valid = x_valid.copy()
tmp_x_test  = x_test.copy()
```

```python
from sklearn.model_selection import KFold # KFold일지 StratifiedKFold를 사용해야하는지 고민해보세요.
kf = KFold(n_splits=5, shuffle=True, random_state=2020)
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor

from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_absolute_error
```

```python
#######Lasso 학습

val_scores1 = list()

for i, (trn_idx, val_idx) in enumerate(kf.split(data, label)) :
    x_train, y_train = data.iloc[trn_idx, :], label.iloc[trn_idx,:]
    x_valid, y_valid = data.iloc[val_idx, :], label.iloc[val_idx,:]
    
    # 전처리
    x_train, x_valid, _ = preprocess(x_train, x_valid, test) # _ : 함수에서 값은 나오지만 처리하지는 않겠다는 의미
    
    # 모델 정의
    ls = MultiOutputRegressor(Lasso(random_state=2020))
    
    # 모델 학습
    ls.fit(x_train, y_train)

    # 훈련, 검증 데이터 
    trn_mae = mean_absolute_error(y_train, ls.predict(x_train))
    val_mae = mean_absolute_error(y_valid, ls.predict(x_valid))
    print('{} Fold, train mae : {:.4f}, validation mae : {:.4f}'.format(i, trn_mae, val_mae))
    
    val_scores1.append(val_mae)

#교차 검증
print('Cross Validation Score : {:.4f}%'.format(np.mean(val_scores1)))
```

```python
#######Ridge 학습

val_scores2 = list()

for i, (trn_idx, val_idx) in enumerate(kf.split(data, label)) :
    x_train, y_train = data.iloc[trn_idx, :], label.iloc[trn_idx,:]
    x_valid, y_valid = data.iloc[val_idx, :], label.iloc[val_idx,:]
    
    # 전처리
    x_train, x_valid, _ = preprocess(x_train, x_valid, test) # _ : 함수에서 값은 나오지만 처리하지는 않겠다는 의미
    
    # 모델 정의
    rg = MultiOutputRegressor(Ridge(random_state=2020))
    
    # 모델 학습
    rg.fit(x_train, y_train)

    # 훈련, 검증 데이터 
    trn_mae = mean_absolute_error(y_train, rg.predict(x_train))
    val_mae = mean_absolute_error(y_valid, rg.predict(x_valid))
    print('{} Fold, train mae : {:.4f}, validation mae : {:.4f}'.format(i, trn_mae, val_mae))
    
    val_scores2.append(val_mae)

#교차 검증
print('Cross Validation Score : {:.4f}%'.format(np.mean(val_scores2)))
```

```python
#######Randomforest 학습

val_scores3 = list()

for i, (trn_idx, val_idx) in enumerate(kf.split(data, label)) :
    x_train, y_train = data.iloc[trn_idx, :], label.iloc[trn_idx,:]
    x_valid, y_valid = data.iloc[val_idx, :], label.iloc[val_idx,:]
    
    # 전처리
    x_train, x_valid, _ = preprocess(x_train, x_valid, test) # _ : 함수에서 값은 나오지만 처리하지는 않겠다는 의미
    
    # 모델 정의
    rf = MultiOutputRegressor(RandomForestRegressor(n_estimators=10, max_depth=50, random_state=2020, n_jobs = 6))
    
    # 모델 학습
    rf.fit(x_train, y_train)

    # 훈련, 검증 데이터 
    trn_mae = mean_absolute_error(y_train, rf.predict(x_train))
    val_mae = mean_absolute_error(y_valid, rf.predict(x_valid))
    print('{} Fold, train mae : {:.4f}, validation mae : {:.4f}'.format(i, trn_mae, val_mae))
    
    val_scores3.append(val_mae)

#교차 검증
print('Cross Validation Score : {:.4f}%'.format(np.mean(val_scores3)))
```

```python
#######LGB 학습

val_scores4 = list()

for i, (trn_idx, val_idx) in enumerate(kf.split(data, label)) :
    x_train, y_train = data.iloc[trn_idx, :], label.iloc[trn_idx,:]
    x_valid, y_valid = data.iloc[val_idx, :], label.iloc[val_idx,:]
    
    # 전처리
    x_train, x_valid, _ = preprocess(x_train, x_valid, test) # _ : 함수에서 값은 나오지만 처리하지는 않겠다는 의미
    
    # 모델 정의
    lgb = MultiOutputRegressor(LGBMRegressor(n_estimators=150, max_depth=10, random_state=2020, n_jobs = 6))
    
    # 모델 학습
    lgb.fit(x_train, y_train)

    # 훈련, 검증 데이터 
    trn_mae = mean_absolute_error(y_train, lgb.predict(x_train))
    val_mae = mean_absolute_error(y_valid, lgb.predict(x_valid))
    print('{} Fold, train mae : {:.4f}, validation mae : {:.4f}'.format(i, trn_mae, val_mae))
    
    val_scores4.append(val_mae)

#교차 검증
print('Cross Validation Score : {:.4f}%'.format(np.mean(val_scores4)))
```

```python
####### oof 앙상블

val_scores_fnl = list()
oof_pred = np.zeros((data.shape[0], label.shape[1]))

for i, (trn_idx, val_idx) in enumerate(kf.split(data, label)):
    x_train, y_train = data.iloc[trn_idx, :], label.iloc[trn_idx,:]
    x_valid, y_valid = data.iloc[val_idx, :], label.iloc[val_idx,:]
    
    # 전처리
    x_train, x_valid, x_test = preprocess(x_train, x_valid, test)
    
    # 모델 정의
    lgb = MultiOutputRegressor(LGBMRegressor(n_estimators=150, max_depth=10, random_state=2020, n_jobs = 6))
    
    # 모델 학습
    lgb.fit(x_train, y_train)

    # 훈련, 검증 데이터 MAE 확인
    trn_mae = mean_absolute_error(y_train, lgb.predict(x_train))
    val_mae = mean_absolute_error(y_valid, lgb.predict(x_valid))
    print('{} Fold, train MAE : {:.4f}4, validation MAE : {:.4f}'.format(i, trn_mae, val_mae))
    
    val_scores_fnl.append(val_mae)
    
    # MAE니까 Fold 전체의 평균
    oof_pred += lgb.predict(x_test) / 5
    

# 교차 검증 
print('Cross Validation Score : {:.4f}'.format(np.mean(val_scores_fnl)))
```
